## 1. Objective

To demonstrate the use of the if-elif-else ladder in Python, which allows for checking multiple expressions and executing a block of code as soon as one of the conditions evaluates to True.

---

## 2. Learning Outcomes

- **Multi-way Branching:** Using `elif` (short for else-if) to check multiple specific values of a single variable.  
- **Mutual Exclusivity:** Understanding that once a condition in the ladder is met, Python skips all remaining `elif` and `else` blocks.  
- **Default Case:** Utilizing the `else` block as a "catch-all" for scenarios where none of the specified conditions match.  

---

## 3. Conclusion

The task confirms that since `i` is exactly 20, the program skips the first two checks, executes the `elif (i == 20)` block, and ignores the final `else` statement. This structure is highly efficient for handling categorized data in AI applications.

```python
i = 20

if (i == 10):
    print("i is 10")
elif (i == 15):
    print("i is 15")
elif (i == 20):
    print("i is 20")
else:
    print("i is not present")
```
