## **1. Objective**

The objective of this task is to understand the **different Loss Functions (Cost Functions)** used in **machine learning and deep learning**. The task focuses on how **errors are calculated** and how algorithms **minimize these errors** during training.  

---

## **2. Learning Outcomes**

* **Loss vs. Cost Function:** Learned that the error for a **single training example** is called **Loss**, while the **average error over the dataset** is called the **Cost Function**.  

* **Regression Loss Functions:**  
  - **Mean Squared Error (MSE):** Squares the error, **punishing large errors more**.  
  - **Mean Absolute Error (MAE):** Takes the absolute error, **robust to outliers**.  

* **Classification Loss Functions:**  
  - **Binary Cross-Entropy (Log Loss):** Used for **binary classification (0 or 1)**. Applies **exponential penalty** for errors.  
  - **Categorical Cross-Entropy:** Best suited for **multi-class classification** problems.  

* **Optimization:** The main purpose of a loss function is to **guide Gradient Descent** so that **weights are updated** in a direction where **loss is minimized**.  

---

## **3. Conclusion**

Choosing the **correct loss function** is **critical** for model performance. Using a classification loss for a regression problem will prevent the model from training. Task 59 highlighted that **machine learning is fundamentally an optimization problem**, where the goal is to **bring error as close to zero as possible**.  

```python
import numpy as np

# --- 1. Regression Loss Functions ---

# Mean Squared Error (MSE)

def mse_loss(y_true, y_pred):
    return np.mean(np.square(y_true - y_pred))

# Mean Absolute Error (MAE)

def mae_loss(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))
```

```python
# --- 2. Classification Loss Functions ---

# Binary Cross-Entropy (Log Loss)

def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
# --- 3. Testing the Functions ---

# Sample data for Regression

true_val = np.array([1.0, 2.0, 3.0])
pred_val = np.array([0.8, 2.5, 2.9])

print("--- Regression Metrics ---")
print(f"MSE Loss: {mse_loss(true_val, pred_val):.4f}")
print(f"MAE Loss: {mae_loss(true_val, pred_val):.4f}")

# Sample data for Classification (Probabilities)

true_labels = np.array([1, 0, 1, 1])
pred_probs = np.array([0.9, 0.1, 0.2, 0.8])

print("\n--- Classification Metrics ---")
print(f"Binary Cross-Entropy Loss: {binary_cross_entropy(true_labels, pred_probs):.4f}")
```
