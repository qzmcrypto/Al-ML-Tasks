## **1. Objective**

The objective of this task is to learn the most important Pandas structure, **DataFrame**. A DataFrame is a **2-dimensional data structure**, like a table with rows and columns. We will explore how to **load data** and how to **locate specific rows**.

---

## **2. Learning Outcomes**

* **DataFrame Structure:** Understanding that a DataFrame works like a table where **multi-dimensional data** can be stored.

* **Locate Row (`loc`):** Learning how to use the `loc` attribute to access specified row(s).

* **Named Indexes:** Assigning custom row names (e.g., "day1", "day2") instead of default integer indexes.

* **Locate Named Indexes:** Extracting specific row data using the named index.

* **File Loading:** Understanding how to load **external files** (such as CSV) into a DataFrame.

---

## **3. Conclusion**

The **DataFrame** is the most powerful feature of the Pandas library. In AI, datasets are always in **tabular form**, so understanding the DataFrame structure and the `.loc` property is a **fundamental skill for data manipulation**. It allows us to extract desired data from thousands of rows in **seconds**.

```python
import pandas as pd

# --- 1. Create a simple Pandas DataFrame ---

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

# Load data into a DataFrame object:

df = pd.DataFrame(data)
print("--- Basic DataFrame ---")
print(df) 
```

```python
# --- 2. Locate Row ---
# Return row 0 (Returns a Series)

print("--- Row 0 ---")
print(df.loc[0])

# Return row 0 and 1 (Returns a DataFrame)

print("\n--- Row 0 and 1 ---")
print(df.loc[[0, 1]])
```

```python
# --- 3. Named Indexes ---

df_named = pd.DataFrame(data, index = ["day1", "day2", "day3"])
print("--- DataFrame with Named Indexes ---")
print(df_named) 
```

```python
# --- 4. Locate Named Indexes ---

# Return "day2"

print("--- Locate 'day2' ---")
print(df_named.loc["day2"])
```
