## **1. Objective**

The objective of this task is to differentiate between a **Copy** and a **View** in NumPy arrays. We will learn how memory is allocated differently for both, how changes in one affect the other, and how to use the `.base` attribute to verify data ownership.

---

## **2. Learning Outcomes**

* **Data Ownership:**
    * **Copy:** It is a new array that **owns** the data. Changes made to the copy do **not** affect the original array, and vice-versa.
    * **View:** It is just a different perspective of the original array and **does not own** the data. Changes made to the view **will reflect** in the original array.

* **Memory Efficiency:** Understanding that views are faster and save memory because they do not create a duplicate of the data in the system's RAM.

* **The .base Attribute:**
    * If an array owns its data (Copy), `.base` returns `None`.
    * If it is a view, `.base` refers to the **original array object**.

---

## **3. Conclusion**

Understanding the difference between **Copy** and **View** is vital to prevent accidental data modification. In Machine Learning, we often use **views** for slicing large datasets to save memory, but we use **copies** when we need to transform data without altering the original source file or database.

```python
import numpy as np

# --- 1. The COPY Method ---
# Copy owns the data

arr = np.array([1, 2, 3, 4, 5])
x = arr.copy()

# Changing original doesn't affect copy

arr[0] = 42

print("--- COPY ---")
print("Original Array (modified):", arr)
print("Copy Array (unchanged):", x)
```

```python
# --- 2. The VIEW Method ---
# View does not own the data

arr_orig = np.array([1, 2, 3, 4, 5])
y = arr_orig.view()

# Changing original DOES affect the view

arr_orig[0] = 42

print("--- VIEW ---")
print("Original Array (modified):", arr_orig)
print("View Array (also modified):", y)

# Changing the view also affects the original

y[1] = 31
print("Original after modifying View:", arr_orig)
```

```python
# --- 3. Checking Ownership using .base ---
# Base attribute tells us if the array owns the data

arr_base = np.array([1, 2, 3, 4, 5])

copy_arr = arr_base.copy()
view_arr = arr_base.view()

print("--- Ownership Check (.base) ---")
print("Copy base (should be None):", copy_arr.base)
print("View base (should show original array):", view_arr.base)
```

```python
# --- 4. Slicing as a View ---
# Important: Slicing in NumPy returns a view, not a copy!

main_arr = np.array([10, 20, 30, 40])
slice_view = main_arr[1:3]
slice_view[0] = 99

print("--- Slicing Behavior ---")
print("Original after modifying slice:", main_arr)
```
