## **1. Objective**

The objective of this task is to learn how to **Search** for specific values or conditions within a NumPy array. We will cover the `where()` method for general searching and the `searchsorted()` method for performing efficient binary searches on sorted data.

---

## **2. Learning Outcomes**

- **Conditional Searching:** Using `np.where()` to find the indices of elements that satisfy a specific condition (e.g., finding all even numbers or searching for a specific value like `x == 4`).

- **Binary Search:** Utilizing `np.searchsorted()` on sorted arrays to find the appropriate insertion point for a value to maintain the existing order.

- **Search Direction:** Learning how to control the search side (`side='left'` or `side='right'`) in sorted arrays to determine where the index should be placed.

- **Batch Searching:** Searching for multiple values simultaneously and receiving an array of their respective insertion indices for high-speed data processing.

---

## **3. Conclusion**

Searching is a vital component of data analysis. In **Machine Learning**, `np.where()` is frequently used to filter datasets, identify outliers, or create mask arrays for data cleaning. Understanding `searchsorted()` is equally important for maintaining sorted data structures, which is a key requirement for many optimized algorithms and efficient data retrieval.

```python
import numpy as np

# --- 1. Basic Searching with where() ---

arr = np.array([1, 2, 3, 4, 5, 4, 4])

# Find the indexes where the value is 4

x = np.where(arr == 4)
print("Indices where value is 4:", x)

# Find indexes where values are even

arr_seq = np.array([1, 2, 3, 4, 5, 6, 7, 8])
even_idx = np.where(arr_seq % 2 == 0)
print("Indices of even numbers:", even_idx)
```

```python
# --- 2. Searching in Sorted Arrays (searchsorted) ---
# searchsorted finds the index where a value should be inserted

arr_sorted = np.array([6, 7, 8, 9])

# Find index where 7 should be inserted (Default side='left')

idx_left = np.searchsorted(arr_sorted, 7)
print("Insertion index for 7 (left):", idx_left)

# Find index where 7 should be inserted (side='right')

idx_right = np.searchsorted(arr_sorted, 7, side='right')
print("Insertion index for 7 (right):", idx_right)
```

```python
# --- 3. Multiple Value Search ---

arr_base = np.array([1, 3, 5, 7])

# Find insertion indices for multiple values: 2, 4, and 6

multi_idx = np.searchsorted(arr_base, [2, 4, 6])
print("Insertion indices for [2, 4, 6]:", multi_idx)
```
