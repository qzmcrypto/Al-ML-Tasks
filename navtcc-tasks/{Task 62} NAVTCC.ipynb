## **2. Learning Outcomes**

* **What is NLP?**  
  NLP is a field of **computer science and linguistics** that gives computers the ability to **process and analyze human language**.  

* **Tokenization:** Splitting text into **smaller units (tokens)**.  
  - **Word Tokenization:** Splitting sentences into individual words.  
  - **Sentence Tokenization:** Splitting paragraphs into separate sentences.  

* **Stop Words Removal:** Removing common words (e.g., 'the', 'is', 'in') that **do not add significant meaning** for classification.  

* **Stemming & Lemmatization:** Reducing words to their **root form**.  
  - **Stemming:** Cuts off word suffixes (e.g., 'running' → 'run').  
  - **Lemmatization:** Uses dictionary to extract the correct root word.  

* **Visualizing Text:** Using **WordCloud** to see the **most frequently used words** in the text.

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Downloading essential NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# --- 1. Tokenization Example ---

data = "Natural Language Processing is fun. Learning with NAVTTC is even better!"

# Word Tokenization

words = word_tokenize(data)
print("Word Tokens:")
print(words)

# Sentence Tokenization

sentences = sent_tokenize(data)
print("\nSentence Tokens:")
print(sentences)
```

```python
# --- 2. Stop Words Removal ---

stop_words = set(stopwords.words('english'))
words_filtered = [w for w in words if not w.lower() in stop_words]

print("Original Words count:", len(words))
print("\nFiltered Words count (without stop words):", len(words_filtered))
print("\nFiltered Words:", words_filtered)
```

```python
# --- 3. Stemming Example ---

ps = PorterStemmer()
example_words = ["coding", "coder", "coded", "codes"]

print("Stemming Results:")
for w in example_words:
    print(f"{w} -> {ps.stem(w)}")
```
