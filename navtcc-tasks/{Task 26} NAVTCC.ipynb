## **1. Objective**

The objective of this task is to learn how to manage runtime errors in Python using **Exception Handling**. Instead of letting a program crash, we use `try...except` blocks to catch errors, ensure smooth execution, and provide meaningful feedback to the user.

---

## **2. Learning Outcomes**

- **Error Resilience:** Understanding that exceptions (like dividing by zero or invalid indices) can be caught to prevent program termination.

- **The Try-Except Block:** Learning to wrap "risky" code in a `try` block and handle specific errors in the `except` block.

- **Specific Exception Catching:** Handling different errors (e.g., `ZeroDivisionError`, `IndexError`, `ValueError`) with separate logic.

- **Control Flow Keywords:**
    - **`else`:** Executed only if **no** exception occurred in the `try` block.
    - **`finally`:** Executed **regardless** of whether an exception occurred (useful for cleanup tasks like closing files).

- **Assertion:** Using `assert` to enforce specific conditions within the code.

---

## **3. Conclusion**

Exception handling is a critical skill for building robust AI systems. When dealing with large datasets or real-time user input, errors are inevitable. By mastering `try...except...finally` blocks, we can write code that handles data inconsistencies or hardware issues gracefully without losing progress.

```python
# --- 1. Basic try...except Block ---
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
    print(result)
except ZeroDivisionError:
    print("Error: Denominator cannot be 0.")


# --- 2. Catching Specific Exceptions ---
try:
    even_numbers = [2, 4, 6, 8]
    # Trying to access index 5 which doesn't exist
    print(even_numbers[5])
except ZeroDivisionError:
    print("Denominator cannot be 0.")
except IndexError:
    print("\nError: Index Out of Bound.")


# --- 3. try with else Clause ---
# The else block runs only if the try block succeeds
try:
    num = int(input("\nEnter an even number for reciprocal calculation: "))
    assert num % 2 == 0
except ValueError:
    print("Error: Please enter a valid integer.")
except AssertionError:
    print("Error: Not an even number!")
except ZeroDivisionError:
    print("Error: Cannot calculate reciprocal of 0.")
else:
    reciprocal = 1 / num
    print(f"The reciprocal is: {reciprocal}")


# --- 4. try...except...finally Block ---
# Finally block always runs, no matter what
try:
    print("\nAttempting division...")
    val = 10 / 0
except:
    print("Catching the error...")
finally:
    print("Execution Finished: This finally block always runs.")


# --- 5. Handling Multiple Exceptions in one block ---
try:
    # Simulating a mixed error scenario
    data = {"name": "Qalb E Zahra"}
    print(data["age"])
except (KeyError, TypeError) as e:
    print(f"\nSpecific Error Caught: {e}")
```
