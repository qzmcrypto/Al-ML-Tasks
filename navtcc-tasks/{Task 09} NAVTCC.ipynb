## 1. Objective

To solve a quadratic equation of the form `ax^2 + bx + c = 0` by calculating the discriminant and finding the roots using the quadratic formula.

---

## 2. Learning Outcomes

- **Mathematical Constants:** Defining the coefficients `a`, `b`, and `c`.  
- **Discriminant Calculation:** Implementing the formula `d = b^2 - 4ac`.  
- **Quadratic Formula:** Using `cmath.sqrt()` to handle both real and complex roots in the formula `x = (-b ± √d) / 2a`.  
- **Advanced Output:** Using indexed formatting `{0}` and `{1}` to display multiple solutions.  

---

## 3. Conclusion

The task confirms Python's efficiency in solving algebraic equations, a core requirement for understanding optimization and loss functions in Machine Learning.

```python
import cmath

a = 1
b = 5
c = 6

d = (b**2) - (4*a*c)

sol1 = (-b - cmath.sqrt(d)) / (2*a)
sol2 = (-b + cmath.sqrt(d)) / (2*a)

print('The solutions are {0} and {1}'.format(sol1, sol2))
```
