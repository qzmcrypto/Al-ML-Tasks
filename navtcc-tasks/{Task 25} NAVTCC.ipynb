## **1. Objective**

The objective of this task is to understand and implement **Python Dictionaries**. Dictionaries are collections that store data in **key-value pairs**. This task covers creating dictionaries, accessing values, modifying entries, and using built-in methods for data management.

---

## **2. Learning Outcomes**

* **Key-Value Structure:** Understanding that every item consists of a unique, immutable **key** and a corresponding **value**.

* **Data Access:** Learning to retrieve values using keys (e.g., `dict['key']`) or the `get()` method.

* **Mutability:** * **Adding/Changing:** Updating existing values or adding new pairs using assignment.
    * **Removing:** Using `del`, `pop()`, or `clear()` to manage memory and data.

* **Dictionary Methods:** Utilizing functions like `keys()`, `values()`, `items()`, and `update()` to manipulate datasets efficiently.

* **Membership Testing:** Checking for the presence of keys using `in` and `not in`.

## **3. Conclusion**

Dictionaries are incredibly powerful for mapping related data, such as database records or **JSON objects**. In the field of Artificial Intelligence, they are frequently used for creating feature maps, storing hyper-parameters, and managing lookup tables for high-speed data retrieval.

```python
# --- 1. Creating a Dictionary ---
# Key: Country, Value: Capital

country_capitals = {
  "Germany": "Berlin", 
  "Canada": "Ottawa", 
  "England": "London"
}
print("Initial Dictionary:", country_capitals)
```

```python
# --- 2. Accessing and Modifying Items ---
# Accessing value using a key

print("Capital of Germany:", country_capitals["Germany"])

# Adding a new item

country_capitals["Italy"] = "Rome"
print("\nAfter adding Italy:", country_capitals)

# Changing an existing value

country_capitals["Italy"] = "Naples" # Correcting value
print("\nAfter updating Italy's capital:", country_capitals)
```

```python
# --- 3. Removing Items ---
# Removing a specific item using 'del'

del country_capitals["Germany"]
print("After deleting Germany:", country_capitals)

# Finding length of the dictionary
print("\nTotal countries in dictionary:", len(country_capitals))
```

```python
# --- 4. Iterating Through a Dictionary ---

print("Iterating through the dictionary:")
for country in country_capitals:
    capital = country_capitals[country]
    print(f"Country: {country}, Capital: {capital}")
```

```python
# --- 5. Dictionary Membership and Methods ---

file_types = {
    ".txt": "Text File",
    ".pdf": "PDF Document",
}

# Testing if a key exists

print("Is '.pdf' in file_types?", ".pdf" in file_types)
print("\nIs '.jpg' in file_types?", ".jpg" in file_types)

# Using .keys() and .values()

print("\nDictionary Keys:", list(file_types.keys()))
print("\nDictionary Values:", list(file_types.values()))
```
