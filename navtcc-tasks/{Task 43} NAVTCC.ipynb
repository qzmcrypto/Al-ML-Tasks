## **1. Objective**

The objective of this task is to understand the basic Pandas structure, **Series**. A Series is a **one-dimensional array** that works like a column in a table and can hold data of any type.

---

## **2. Learning Outcomes**

* **Pandas Series:** Learning what a Series is and how to create it from a list.

* **Labels:** Understanding default indexing (0, 1, 2…) and using it to access data.

* **Create Labels:** Using the `index` argument to assign custom labels (e.g., x, y, z).

* **Key/Value Objects:** Converting a dictionary into a Series, where keys automatically become labels.

* **DataFrames Preview:** Understanding that a Series is a column, while a DataFrame is a full multi-dimensional table.

---

## **3. Conclusion**

**Pandas Series** form the foundation of data manipulation. In AI, when we work on a feature or target variable, it usually exists as a **Series**. Understanding labels and key-value mapping is crucial for **data cleaning and selection**.

```python
import pandas as pd

# --- 1. Simple Series from a list ---

a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)

# Return the first value

print(myvar[0])
```

```python
# --- 2. Create Labels ---

myvar_labels = pd.Series(a, index = ["x", "y", "z"])
print(myvar_labels)

# Return the value of "y"

print(myvar_labels["y"])
```

```python
# --- 3. Series from Dictionary ---

calories = {"day1": 420, "day2": 380, "day3": 390}
myvar_dict = pd.Series(calories)
print(myvar_dict)
```

```python
# --- 4. Series using only specific items ---

myvar_select = pd.Series(calories, index = ["day1", "day2"])
print(myvar_select)
```

```python
# --- 5. Creating a DataFrame from two Series ---

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
myvar_df = pd.DataFrame(data)
print(myvar_df)
```
