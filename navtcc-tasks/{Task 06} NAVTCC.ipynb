## 1. Objective

To calculate the square root of a complex number using Python's built-in `cmath` (complex math) module and format the real and imaginary parts of the result.

---

## 2. Learning Outcomes

- **Module Importing:** Learning to use `import cmath` for advanced mathematical functions beyond standard real numbers.  
- **Complex Data Types:** Working with the `j` notation to represent imaginary parts in Python.  
- **Component Access:** Accessing the `.real` and `.imag` attributes of a complex number object.  
- **Advanced Formatting:** Using index-based formatting `{1:0.3f}` to control decimal precision for multiple values.  

---

## 3. Conclusion

The task confirms that Python is highly capable of scientific computing, handling complex number algebra which is essential for advanced AI algorithms and signal processing.

```python
import cmath

num = 1 + 2j

num_sqrt = cmath.sqrt(num)

print('The square root of {0} is {1:0.3f}+{2:0.3f}j'.format(num, num_sqrt.real, num_sqrt.imag))
```
