## 1. Objective

To implement a Python program that converts temperature from Celsius to Fahrenheit using a standard mathematical formula.

---

## 2. Learning Outcomes

- **Mathematical Operations:** Applying the conversion formula `F = (C × 1.8) + 32`.  
- **Variable Initialization:** Assigning floating-point values to represent temperature.  
- **Precision Formatting:** Using `%0.1f` to display the result with one decimal point.  

---

## 3. Conclusion

The task confirms that Python can effectively handle unit conversions and display results with specific formatting, which is a foundational skill for data preprocessing in AI.

```python
celsius = 37.5

fahrenheit = (celsius * 1.8) + 32

print('%0.1f degree Celsius is equal to %0.1f degree Fahrenheit' % (celsius, fahrenheit))
```
