```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
```

```python
# --- 1. Loading Data ---

url = "https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv"
pima = pd.read_csv(url)
```

```python
# --- 2. Feature Selection ---

feature_cols = ['Pregnancies', 'Insulin', 'BMI', 'Age', 'Glucose', 'BloodPressure', 'DiabetesPedigreeFunction']
X = pima[feature_cols]
y = pima.Outcome
```

```python
# --- 3. Splitting Data ---
# 70% training aur 30% test data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# --- 4. Building Decision Tree Model (Unoptimized) ---

clf = DecisionTreeClassifier(random_state=1)
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("--- Step 4 Results ---")
print("Standard Accuracy:", round(metrics.accuracy_score(y_test, y_pred), 4))
```

```python
# --- 5. Optimizing Decision Tree Performance (Task 53 Main Goal) ---
# Optimization using 'entropy' and 'max_depth=3'

clf_optimized = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=1)

clf_optimized = clf_optimized.fit(X_train, y_train)

y_pred_opt = clf_optimized.predict(X_test)
```

```python
# --- 6. Final Evaluation ---

print("\n--- Step 6 Results ---")
print("Optimized Accuracy (max_depth=3):", round(metrics.accuracy_score(y_test, y_pred_opt), 4))
```

```python
# Optional: Checking a manual prediction
# Dummy data: [Pregnancies, Insulin, BMI, Age, Glucose, BloodPressure, Pedigree]

sample_patient = pd.DataFrame([[2, 100, 25.5, 30, 120, 80, 0.5]], columns=feature_cols)
prediction = clf_optimized.predict(sample_patient)

print(f"Prediction: {prediction[0]}")
```
