## **1. Objective**

The objective of this task is to understand the **working and architecture** of **Feedforward Neural Networks (FNNs)**. These networks form the **foundation of deep learning** and process data to **recognize patterns**.  

---

## **2. Learning Outcomes**

* **The Mesh Architecture:** In a feedforward network, data flows from **input**, through **hidden layers**, to the **output layer**. There are **no loops or cycles**.  

* **Weighted Sum & Bias:** Each neuron multiplies inputs by **weights** and adds a **bias**:  

  $$
  \text{Weighted Sum} = (x_1 \cdot w_1) + (x_2 \cdot w_2) + (x_3 \cdot w_3) + \text{Bias}
  $$  

* **Activation Cheatsheet:** Different problems require different activation functions:  
  - **ReLU:** For hidden layers (sets negative values to zero).  
  - **Sigmoid:** For binary classification (outputs between 0 and 1).  
  - **Softmax:** For multi-class classification (outputs probabilities across classes).  

* **Loss Function & Optimization:** Model error is measured using a **loss function**, and **Gradient Descent** updates weights to **minimize error**.  

---

## **3. Conclusion**

**Feedforward Neural Networks** are the **first and highly effective deep learning algorithm**. Frameworks like **TensorFlow** make implementing these networks easy. Understanding FNNs is essential for learning **Convolutional Neural Networks (CNNs)** for images and **Recurrent Neural Networks (RNNs)** for text.  

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Input
import numpy as np
```

```python
# --- 1. Data Setup ---
# Simple dummy data (e.g., predicting weather based on 3 inputs)
# x1: day/night, x2: temp, x3: month

X_train = np.array([[0, 12, 11], [1, 25, 6], [0, 30, 7], [1, 10, 12]], dtype=float)
y_train = np.array([1, 0, 0, 1], dtype=float)
```

```python
# --- 2. Building the FNN Model (Warning-Free Version) ---

model = Sequential()

model.add(Input(shape=(3,))) 

model.add(Dense(8, activation='relu'))

model.add(Dense(4, activation='relu'))

model.add(Dense(1, activation='sigmoid'))
```

```python
# --- 3. Compiling the Model ---
# Using 'adam' optimizer and 'binary_crossentropy' for loss

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# --- 4. Training the Model ---

print("Starting Training...")
model.fit(X_train, y_train, epochs=100, verbose=0)
```

```python
# --- 5. Making a Prediction ---
# Test data: (0, 12, 11) - similar to the example in the article

test_data = np.array([[0, 12, 11]])
prediction = model.predict(test_data)

print(f"Prediction for inputs (0, 12, 11): {'Rainy' if prediction > 0.5 else 'Sunny'}")
print(f"\nRaw Output Probability: {prediction[0][0]:.4f}")
```
