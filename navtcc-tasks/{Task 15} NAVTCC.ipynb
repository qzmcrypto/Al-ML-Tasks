## **3. Conclusion**

The completion of this task confirms that `for` loops are essential for repetitive tasks in Python. By using control statements like `break` and `continue`, we can fine-tune the flow of data processing, which is a fundamental requirement for data manipulation in Machine Learning.

```python
# Iterating over a list

s = ["Geeks", "for", "Geeks"]
print("List Iteration:")
for i in s:
    print(i)

print("\nString Iteration:")

# Iterating over characters of a string

s = "Geeks"
for i in s:
    print(i)
```

```python
# Using range(start, stop, step)

print("Range Function (0 to 8 with step 2):")
for i in range(0, 10, 2):
    print(i)

# Continue Statement: Prints all letters except 'e' and 's'

print("\nUsing Continue:")
for i in 'geeksforgeeks':
    if i == 'e' or i == 's':
        continue
    print(i, end=" ")

# Break Statement: Stops at first 'e' or 's'

print("\n\nUsing Break:")
for i in 'geeksforgeeks':
    if i == 'e' or i == 's':
        break
    print(i)
```

```python
# Using Enumerate to get index and value

print("Using Enumerate:")
li = ["eat", "sleep", "repeat"]
for index, value in enumerate(li):
    print(index, value)

# Nested For Loops

print("\nNested Loops (i and j):")
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j)
```
