## **1. Objective**

The objective of this task is to implement the `while` loop in Python to execute a block of code repeatedly as long as a specified condition is met. We will explore the loop's syntax, infinite loops, and how control statements like `break`, `continue`, and `pass` modify its behavior.

---

## **2. Learning Outcomes**

- **Condition-Based Looping:**  
  Understanding that a `while` loop runs only as long as its expression evaluates to `True`.  

- **Loop Termination:**  
  Identifying how a loop ends when the condition becomes `False`.  

- **Control Statements:**  

  - **Continue:**  
    Returns control to the beginning of the loop.  

  - **Break:**  
    Brings control out of the loop immediately.  

  - **Pass:**  
    Used to create an empty loop structure.  

- **Else Clause:**  
  Learning that the `else` block executes only if the loop completes normally without hitting a `break`.  

---

## **3. Conclusion**

This task confirms that `while` loops are dynamic tools for situations where the number of iterations is not known beforehand. Mastery of these loops and their control statements is essential for building robust logic in AI and Machine Learning algorithms.

```python
# --- Example 1: Basic While Loop ---

print("Basic While Loop (Counter):")
count = 0
while count < 3:
    count = count + 1
    print("Hello Geek")
```

```python
# --- Example 2: While loop with Continue ---

# Skips printing 'e' and 's'

print("Using Continue Statement:")
i = 0
a = 'geeksforgeeks'
while i < len(a):
    if a[i] == 'e' or a[i] == 's':
        i += 1
        continue
    print(a[i], end=" ")
    i += 1
```

```python
# --- Example 3: While loop with Break ---

# Stops the loop at the first 'e' or 's'

print("Using Break Statement:")
i = 0
while i < len(a):
    if a[i] == 'e' or a[i] == 's':
        break
    print(a[i])
    i += 1
```

```python
# --- Example 4: While loop with Else ---

print("Using While-Else (No Break):")
n = 1
while n < 4:
    print(n)
    n += 1
else:
    print("Loop finished (Condition became False)")
```
