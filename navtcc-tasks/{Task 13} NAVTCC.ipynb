## 1. Objective

To demonstrate the use of Nested If statements, where one conditional block is placed inside another to perform multi-level logical checks.

---

## 2. Learning Outcomes

- **Hierarchy of Logic:** Understanding that nested if blocks only execute if the parent (outer) if condition is True.  
- **Multi-stage Validation:** Performing a specific check (`i == 10`) followed by sub-checks (`i < 15` and `i < 12`).  
- **Indentation Levels:** Managing multiple levels of indentation to keep track of which else belongs to which if.  

---

## 3. Conclusion

The task confirms that since `i` is exactly 10, the outer block is entered. Both nested if conditions (`10 < 15` and `10 < 12`) are also True, resulting in multiple successful print statements within a single logical branch.

```python
i = 10

if (i == 10):
    print("i is 10")
    
    if (i < 15):
        print("i is smaller than 15")
        
    if (i < 12):
        print("i is smaller than 12 too")
    else:
        print("i is greater than 15")
```
