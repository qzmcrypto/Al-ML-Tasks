```python
# --- 1. Basic Function and Calling ---

def welcome_fun():
    print("Welcome to GFG")

welcome_fun()
```

```python
# --- 2. Function with Arguments and Return ---

def evenOdd(x):
    if (x % 2 == 0):
        return "Even"
    else:
        return "Odd"

print(f"Number 16 is: {evenOdd(16)}")
```

```python
# --- 3. Default and Keyword Arguments ---

def student_info(fname, lname, age=20):
    print(f"Name: {fname} {lname}, Age: {age}")

student_info(lname='Practice', fname='Geeks')
```

```python
# --- 4. Arbitrary Arguments (*args and **kwargs) ---

def myFun(*args, **kwargs):
    print("\nNon-Keyword (*args):", args)
    print("Keyword (**kwargs):", kwargs)

myFun('Hey', 'Welcome', first='Geeks', last='Geeks')
```

```python
# --- 5. Anonymous Function (Lambda) ---

cube = lambda x : x*x*x  
print(f"\nCube of 7 using Lambda: {cube(7)}")
```

```python
# --- 6. Recursive Function (Factorial) ---

def factorial(n):
    if n == 0:  
        return 1
    else:
        return n * factorial(n - 1) 
      
print(f"Factorial of 4: {factorial(4)}")
```

```python
# --- 7. Pass by Reference (Mutable vs Immutable) ---

def modify_data(lst, num):
    lst[0] = 99 
    num = 99    

my_list = [10, 20]
my_num = 10
modify_data(my_list, my_num)
print(f"\nModified List: {my_list}")
print(f"Number remains: {my_num}")
```
