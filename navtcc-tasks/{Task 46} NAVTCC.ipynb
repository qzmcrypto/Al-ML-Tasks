## **2. Learning Outcomes**

## **3. Conclusion**

Analyzing DataFrames is a **critical step** in the **Exploratory Data Analysis (EDA)** phase of any Machine Learning project. Methods like `head()` and `info()` allow us to verify if data has loaded correctly and to spot inconsistencies, such as missing values in a column. Cleaning these null values ensures that our **AI models** are trained on **high-quality, complete data**.

```python
import pandas as pd

# Load the CSV file into a DataFrame
# (Ensure 'data.csv' is in your directory or use a URL)

df = pd.read_csv('data.csv')

# --- 1. Viewing the Top Rows ---

print("--- First 10 Rows ---")
print(df.head(10))

# Default head() returns 5 rows

print("\n--- Default head() (Top 5) ---")
print(df.head())
```

```python
# --- 2. Viewing the Bottom Rows ---

print("--- Last 5 Rows ---")
print(df.tail())
```

```python
# --- 3. Getting Information About the Data ---

print("--- DataFrame Information ---")
df.info()

# Note: In the info() output, look for columns where the 
# 'Non-Null Count' is less than the total 'Entries' to find missing data.
```
