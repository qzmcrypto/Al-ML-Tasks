## **2. Learning Outcomes**

* **Raising Custom Errors:** Using the `raise` keyword to trigger our user-defined exceptions based on specific conditions.

## **3. Conclusion**

Custom exceptions allow developers to write more readable and maintainable code. Instead of using generic errors like `ValueError`, we can use specific names like `InvalidAgeException` or `SalaryNotInRangeError`. This is especially useful in complex AI pipelines where we need to identify exactly which part of the data preprocessing or model validation failed.

```python
# --- 1. Basic Custom Exception ---
# Defining a custom exception class

class InvalidAgeException(Exception):
    """Raised when the input value is less than 18"""
    pass

# Using the custom exception

min_voting_age = 18
try:
    input_num = int(input("Enter your age to check eligibility: "))
    if input_num < min_voting_age:
        raise InvalidAgeException
    else:
        print("You are eligible to vote!")
except InvalidAgeException:
    print("Exception occurred: Invalid Age for voting.")
except ValueError:
    print("Error: Please enter a numeric value for age.")
```

```python
# --- 2. Customized Exception Class with Arguments ---

class SalaryNotInRangeError(Exception):
    """Exception raised for errors in the input salary.

    Attributes:
        salary -- input salary which caused the error
        message -- explanation of the error
    """

    def __init__(self, salary, message="Salary is not in (5000, 15000) range"):
        self.salary = salary
        self.message = message
        super().__init__(self.message)

# Testing the advanced custom exception

try:
    salary_amount = int(input("\nEnter salary amount: "))
    if not 5000 < salary_amount < 15000:
        raise SalaryNotInRangeError(salary_amount)
    else:
        print("Salary is within the acceptable range.")
except SalaryNotInRangeError as e:
    print(f"Custom Error: {e.message} (Input was: {e.salary})")
except ValueError:
    print("Error: Invalid salary input.")
```

```python
# --- 3. Example of Multiple Custom Exceptions ---

class NegativeNumberError(Exception):
    pass

def check_positive(number):
    if number < 0:
        raise NegativeNumberError("Negative numbers are not allowed!")
    return number

try:
    val = check_positive(-5)
except NegativeNumberError as e:
    print(f"Caught specific error: {e}")
```
