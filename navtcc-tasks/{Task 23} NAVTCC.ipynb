## **1. Objective**

The objective of this task is to explore **Python Strings**, which are sequences of characters. We will cover string creation, indexing, slicing, and the concept of immutability. Additionally, we will look into common string operations like concatenation, multiline strings, and modern formatting using **f-strings**.

---

## **2. Learning Outcomes**

* **String Representation:** Using single (`'`), double (`"`), or triple (`"""`) quotes to define text data.

* **Character Access:** * **Indexing:** Accessing specific characters (e.g., `str[0]`).  
    * **Slicing:** Extracting a range of characters (e.g., `str[1:4]`).

* **Immutability:** Understanding that individual characters in a string **cannot** be changed after the string is created.

* **String Operations:** * **Concatenation:** Joining strings using the `+` operator.  
    * **Membership:** Checking for substrings using the `in` keyword.  
    * **Length:** Calculating the number of characters using `len()`.  
    * **Escape Sequences:** Using `\` to include special characters like quotes or newlines inside a string.  
    * **f-Strings:** Implementing concise and readable string formatting.

---

## **3. Conclusion**

Strings are fundamental to almost every Python application, from simple user prompts to complex text processing in **Natural Language Processing (NLP)**. Understanding their immutable nature and mastering slicing techniques is vital for efficient data manipulation in AI projects.

```python
# --- 1. Creating and Printing Strings ---

name = "Python"
message = 'I love Python.'
print(f"Name: {name}")
print(f"Message: {message}")

# Multiline String

multiline = """
Never gonna give you up
Never gonna let you down
"""
print("Multiline String:", multiline)
```

```python
# --- 2. Accessing Characters (Indexing & Slicing) ---

greet = 'Hello'

# Positive and Negative Indexing

print(f"1st character of '{greet}':", greet[0])
print(f"\n4th last character of '{greet}':", greet[-4])

# Slicing
# Accessing character from 1st index to 3rd index (4 is excluded)

print("Slicing [1:4]:", greet[1:4]) 
```

```python
# --- 3. String Immutability ---

msg = "Hola Amigos"
print(f"Original Message: {msg}")

# Trying to change a character will raise a TypeError

try:
    msg[0] = 'H'
except TypeError as e:
    print(f"\nError: {e} (Strings are immutable!)")

# However, we can reassign the whole variable

msg = "Hello Friends"
print("Reassigned Message:", msg)
```

```python
# --- 4. String Operations ---

str1 = "Hello"
str2 = "World"

# Concatenation

result = str1 + " " + str2
print("Concatenated String:", result)

# Membership Test

print("\n'a' in 'program'?", 'a' in 'program')
print("'z' in 'program'?", 'z' in 'program')

# Finding Length

print("Length of 'Hello':", len(str1))
```

```python
# --- 5. Escape Sequences and Methods ---
# Using backslash to escape quotes

example = "He said, \"What's there?\""
print("Escaped String:", example)

# Using common methods

text = "python programming"
print("\nUppercase:", text.upper())
print("Replace 'python' with 'AI':", text.replace("python", "AI"))
```
