## **1. Objective**

The objective of this task is to understand how **NumPy** handles different **Data Types**. NumPy provides more granular control over data types than standard Python, allowing for memory optimization. We will learn how to check the `dtype` of an array, define specific types during creation, and convert existing types using the `astype()` method.

---

## **2. Learning Outcomes**

- **Common NumPy Symbols:** Understanding the shorthand characters used to represent data types:
    - `i` - integer, `f` - float, `b` - boolean, `u` - unsigned integer.
    - `S` - string, `U` - unicode string, `M` - datetime.

- **Checking Types:** Using the `.dtype` property to inspect the data type of any NumPy array.

- **Explicit Type Definition:** Using the `dtype` argument in `np.array()` to force a specific format (e.g., `dtype='i4'` for a 4-byte integer).

- **Type Conversion (astype):** Learning that the best way to change a type is to create a copy using the `astype()` method.

- **Handling Errors:** Recognizing that a `ValueError` occurs if you try to cast incompatible data (like a string `'a'` into an integer).

---

## **3. Conclusion**

Managing data types is crucial for **memory efficiency** in Artificial Intelligence. For instance, using `int8` instead of `int64` for small numbers can significantly reduce the memory footprint of large datasets. Knowing how to convert types (e.g., changing integers to floats for mathematical operations) ensures that your models perform accurately and efficiently.

```python
import numpy as np

# --- 1. Checking Data Types ---

arr1 = np.array([1, 2, 3, 4])
arr2 = np.array(['apple', 'banana', 'cherry'])

print(f"Type of arr1: {arr1.dtype}")
print(f"Type of arr2: {arr2.dtype}")
```

```python
# --- 2. Creating Arrays with Defined Types ---
# Creating a string array from integers

arr_string = np.array([1, 2, 3, 4], dtype='S')
print("String Array:", arr_string)
print("Dtype:", arr_string.dtype)

# Creating a 4-byte integer array

arr_int4 = np.array([1, 2, 3, 4], dtype='i4')
print("\n4-Byte Integer Array:", arr_int4)
print("Dtype:", arr_int4.dtype)
```

```python
# --- 3. Converting Data Types (astype) ---

arr_float = np.array([1.1, 2.1, 3.1])

# Method 1: Using shorthand 'i'

newarr_i = arr_float.astype('i')
print("Converted to int ('i'):", newarr_i)

# Method 2: Using direct Python type int

newarr_int = arr_float.astype(int)
print("Converted to int (int):", newarr_int)

# Converting to Boolean

arr_bool_test = np.array([1, 0, 3])
newarr_bool = arr_bool_test.astype(bool)
print("Converted to Boolean:", newarr_bool)
```

```python
# --- 4. Handling Conversion Errors ---

try:
    err_arr = np.array(['a', '2', '3'], dtype='i')
except ValueError as e:
    print("Caught Expected Error:", e)
```
