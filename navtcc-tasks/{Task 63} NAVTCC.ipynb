## **1. Objective**

The objective of this task is to understand the **working and architecture of Convolutional Neural Networks (CNN)** in **simple terms**. CNN is a deep learning architecture designed to **automatically extract features** (like edges, shapes, and faces) from images.  

---

## **2. Learning Outcomes**

* **Why CNN for Images?**  
  Normal neural networks are **not sufficient for images** because they cannot capture **spatial relationships between pixels**, whereas CNN preserves the **image structure**.  

* **The Convolution Layer (The Filter):**  
  CNN uses **filters or kernels** that slide over the image to **extract important features**.  

* **Pooling Layer (Downsampling):**  
  This layer **reduces image dimensions** for easier computation while retaining **critical information** (features). **Max Pooling** is commonly used.  

* **Flattening & Fully Connected Layer:**  
  Extracted features are converted into a **long vector** and fed into a **standard neural network** for **final classification** (e.g., Cat or Dog).  

* **Padding & Stride:**  
  Learn how **zero-padding** protects image edges and **strides** control the movement of filters over the image.  

---

## **3. Conclusion**

CNN has **revolutionized computer vision**. From **face recognition** to **self-driving cars**, CNNs are everywhere. This task demonstrated why it is important to analyze **image data as features instead of raw pixels** in machine learning.  

```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
# --- 1. Load Dataset (CIFAR-10: 60,000 images in 10 classes) ---

(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize pixel values to be between 0 and 1

train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
from tensorflow.keras import layers, models

model = models.Sequential()

# Use Input(shape) instead of input_shape in the Conv2D layer

model.add(layers.Input(shape=(32, 32, 3)))

# Convolutional Layer 1

model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Convolutional Layer 2

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))

# Convolutional Layer 3

model.add(layers.Conv2D(64, (3, 3), activation='relu'))
```

```python
# --- 3. Adding Dense Layers (The Classifier) ---

model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))
```

```python
# --- 4. Compiling and Training ---

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

print("Starting CNN Training (Small sample for demo)...")

# Training for only 2 epochs for demonstration

model.fit(train_images[:5000], train_labels[:5000], epochs=2, validation_data=(test_images[:1000], test_labels[:1000]))
```

```python
# --- 5. Summary ---

print("CNN Model Architecture Summary:")
model.summary()
```
