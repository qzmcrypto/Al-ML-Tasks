## 1. Objective

To calculate the area of a triangle when the lengths of all three sides are known, using Heron's Formula in Python.

---

## 2. Learning Outcomes

- **Mathematical Implementation:** Applying the semi-perimeter formula `s = (a + b + c) / 2`.  
- **Complex Expressions:** Implementing Heron's Formula: `Area = √(s(s-a)(s-b)(s-c))`.  
- **Operational Logic:** Combining addition, division, multiplication, and exponentiation (`** 0.5`) in a single workflow.  

---

## 3. Conclusion

The task confirms Python's ability to handle multi-step geometric calculations accurately, providing a foundation for spatial data processing in computer vision and AI.

```python
a = 5
b = 6
c = 7

s = (a + b + c) / 2 

area = (s * (s - a) * (s - b) * (s - c)) ** 0.5

print('The area of the triangle is %0.2f' % area)
```
