## **1. Objective**

The objective of this task is to understand **Python Sets**, which are unordered collections of unique elements. This task covers set creation, handling duplicate values, adding/removing items, and performing mathematical set operations like Union, Intersection, and Difference.

---

## **2. Learning Outcomes**

- **Uniqueness & Mutability:** Understanding that sets automatically remove duplicates and are mutable, though they contain only hashable (immutable) objects.

- **Unordered Nature:** Recognizing that sets do not support indexing or slicing because they do not maintain a specific order.

- **Set Operations:** - **Union ($A \cup B$):** Combining all unique elements from both sets.  
  - **Intersection ($A \cap B$):** Finding common elements between sets.  
  - **Difference ($A - B$):** Identifying elements present in the first set but not the second.  
  - **Symmetric Difference ($A \Delta B$):** Elements in either set, but not in both.

- **Built-in Methods:** Using `add()`, `update()`, `discard()`, and `len()` for efficient data management.

---

## **3. Conclusion**

Sets are incredibly useful in Python when we need to ensure data uniqueness or perform mathematical comparisons between groups of data. In Data Science and AI, sets are frequently used to filter out duplicate records and manage overlapping features in datasets.

```python
# --- 1. Creating Sets and Handling Duplicates ---

# Set of integers (duplicates are automatically removed)

student_id = {112, 114, 116, 118, 115, 112}
print("Student IDs (Unique):", student_id)

# Creating an empty set (Requires set() function)

empty_set = set()
print(f"Type of empty_set: {type(empty_set)}")
```

```python
# --- 2. Modifying a Set ---

numbers = {21, 34, 54, 12}

# add(): Adding a single element

numbers.add(32)
print("After add(32):", numbers)

# update(): Adding multiple elements from a list

tech_companies = ['apple', 'google']
numbers.update(tech_companies)
print("\nAfter update with list:", numbers)

# discard(): Removing an element

numbers.discard(21)
print("\nAfter discarding 21:", numbers)
```

```python
# --- 3. Mathematical Set Operations ---

A = {1, 3, 5, 7}
B = {1, 2, 4, 6, 7}

# Union (All elements)

print("Union (A | B):", A | B)

# Intersection (Common elements)

print("\nIntersection (A & B):", A & B)

# Difference (Elements in A but not in B)

print("\nDifference (A - B):", A - B)

# Symmetric Difference (Elements not in both)

print("\nSymmetric Difference (A ^ B):", A ^ B)
```

```python
# --- 4. Iteration and Length ---

fruits = {"Apple", "Peach", "Mango"}
print("Number of fruits:", len(fruits))

print("\nIterating through set:")
for fruit in fruits:
    print(f"- {fruit}")
```
