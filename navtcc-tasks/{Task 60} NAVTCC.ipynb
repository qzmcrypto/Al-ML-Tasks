## **2. Learning Outcomes**

* **Tokenization:** Splitting text into **smaller units** like **sentences or words**. This is the **first and critical step** in NLP.  

* **Stop Words Removal:** Removing **common words** (e.g., "the", "is", "at") that **do not add meaningful information**, allowing the model to focus on **important keywords**.  

* **Stemming & Lemmatization:** Reducing words to their **root form**:  
  - **Stemming:** Cuts off word endings (e.g., "running" → "run").  
  - **Lemmatization:** Converts words to their **dictionary form**.  

* **Part of Speech (POS) Tagging:** Assigning each word its **grammatical category** (Noun, Verb, Adjective).  

* **WordCloud:** Visualizing the **most frequent words** in the text data.

## **3. Conclusion**

**NLP** is widely used in **chatbots, translation (e.g., Google Translate), and sentiment analysis**. The **NLTK library** makes text analytics in Python much easier. Task 60 highlighted the **importance of text preprocessing** before training any machine learning model.  

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# Downloading necessary resources
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# --- 1. Sample Text ---

text = "Natural Language Processing is fascinating. Machines are learning to understand human languages!"
```

```python
# --- 2. Tokenization ---

print("Word Tokens:")
words = word_tokenize(text)
print(words)
```

```python
# --- 3. Stop Words Removal ---

stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if not w.lower() in stop_words]
print("Filtered Words (No Stop Words):")
print(filtered_words)
```

```python
# --- 4. Stemming & Lemmatization ---

ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("Stemming vs Lemmatization:")
word_to_test = "learning"
print(f"Original: {word_to_test}")
print(f"Stemmed:  {ps.stem(word_to_test)}")
print(f"Lemmatized: {lemmatizer.lemmatize(word_to_test, pos='v')}")
```

```python
# --- 5. Word Frequency Distribution ---

from nltk.probability import FreqDist
fdist = FreqDist(filtered_words)
print("Most Common Words:")
print(fdist.most_common(2))
```
