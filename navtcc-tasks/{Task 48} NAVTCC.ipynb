## **1. Objective**

The objective of this task is to explore the `corr()` method in Pandas to find **relationships between different columns** in a dataset. We will learn how to **interpret correlation coefficients**, distinguish between good and bad correlations, and understand how these relationships help in **predictive modeling**.

---

## **2. Learning Outcomes**

* **Correlation Method:** Learning how `df.corr()` calculates the relationship between **numeric columns** while ignoring non-numeric data.

* **Correlation Scale:** Understanding that correlation values range from **-1 to 1**, where 1 is a perfect positive relationship and -1 is a perfect negative relationship.

* **Interpreting Results:**  
  * **Perfect Correlation (1.0):** A column’s relationship with itself.  
  * **Good Correlation (0.6 to 0.9):** High predictability (e.g., Duration vs. Calories).  
  * **Bad Correlation (< 0.2):** No predictable relationship (e.g., Duration vs. Maxpulse).

* **Predictive Insights:** Learning that **high correlation** allows us to predict one value based on another, which is a core concept in Machine Learning.

---

## **3. Conclusion**

**Data correlation** is a vital tool for **Feature Selection** in Artificial Intelligence. By using the `corr()` method, we can identify which features (columns) have the strongest impact on our **target variable**. For instance, knowing that "Duration" and "Calories" have a high correlation of **0.922** indicates that "Duration" is a strong predictor for weight loss models. This prevents the use of irrelevant data, making our AI models **more efficient and accurate**.

```python
import pandas as pd

# Load the dataset
# (Ensure 'data.csv' is in your directory or use a sample DataFrame)
df = pd.read_csv('data.csv')
```

```python
# --- 1. Finding Relationships ---

# The corr() method calculates the correlation between columns

correlation_matrix = df.corr(numeric_only=True)

print("--- Correlation Matrix ---")
print(correlation_matrix)
```

```python
# --- 2. Analyzing Specific Relationships ---

# Perfect Correlation
print("Note: Duration vs Duration is always 1.0")

# High Positive Correlation

# (Example: Duration and Calories)

if 'Duration' in df.columns and 'Calories' in df.columns:
    print("\nGood Correlation Example:")
    print(f"Duration vs Calories: {correlation_matrix.loc['Duration', 'Calories']}")

# Low/Bad Correlation

# (Example: Duration and Maxpulse)

if 'Duration' in df.columns and 'Maxpulse' in df.columns:
    print("\nBad Correlation Example:")
    print(f"Duration vs Maxpulse: {correlation_matrix.loc['Duration', 'Maxpulse']}")
```
