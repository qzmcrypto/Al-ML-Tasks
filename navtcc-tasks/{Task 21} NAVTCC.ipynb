## **2. Learning Outcomes**

* **List Fundamentals:** Understanding that lists are ordered, mutable (changeable), and allow duplicate values.

* **Indexing and Slicing:** * **Positive Indexing:** Accessing elements starting from index `0`.  
    * **Negative Indexing:** Accessing elements from the end of the list starting from `-1`.  
    * **Slicing:** Extracting specific portions of a list using the `:` operator.

* **Dynamic Manipulation:** * **Adding Items:** Using `append()`, `insert()`, and `extend()`.  
    * **Changing Items:** Modifying values using specific indices.  
    * **Removing Items:** Using `remove()`, `pop()`, and the `del` statement.

* **List Operations:** Iterating through lists using loops and using the `len()` function to determine list size.

## **3. Conclusion**

```python
# --- 1. Creating a Python List ---

# A list of three elements

ages = [19, 26, 29]
print("Ages List:", ages)

# List containing different data types

student = ['Jack', 32, 'Computer Science', [2, 4]]
print("Student Info:", student)

# Empty list

empty_list = []
print("Empty List:", empty_list)
```

```python
# --- 2. Accessing List Elements ---

languages = ['Python', 'Swift', 'C++']

# Accessing using positive index

print("First element (index 0):", languages[0])
print("Third element (index 2):", languages[2])

# Accessing using negative index

print("\nLast item (index -1):", languages[-1])
print("Third last item (index -3):", languages[-3])
```

```python
# --- 3. Slicing of a List ---

my_list = ['p', 'r', 'o', 'g', 'r', 'a', 'm']

# Elements from index 2 to 4

print("Slicing [2:5]:", my_list[2:5])

# Elements from index 5 to end

print("\nSlicing [5:]:", my_list[5:])

# Elements from start to end (Full list)

print("\nFull list slicing [:]:", my_list[:])
```

```python
# --- 4. Adding and Changing Elements ---

fruits = ['apple', 'banana', 'orange']

# append(): Add to the end

fruits.append('cherry')
print("After append:", fruits)

# insert(): Add at specific index

fruits.insert(2, 'mango')
print("\nAfter insert at index 2:", fruits)

# Changing elements

fruits[0] = 'Blueberry'
print("\nAfter changing first element:", fruits)
```

```python
# --- 5. Removing Elements ---

numbers = [2, 4, 7, 9, 10]

# remove(): Removes specific value

numbers.remove(4)
print("After removing 4:", numbers)

# del: Delete item at specific index

names = ['John', 'Eva', 'Laura', 'Nick']
del names[1]
print("\nAfter del names[1]:", names)

# len(): Finding length of list

print("\nLength of names list:", len(names))
```

```python
# --- 6. Iterating Through a List ---

print("Iterating through fruits list:")
for fruit in fruits:
    print(f"Fruit: {fruit}")
```
