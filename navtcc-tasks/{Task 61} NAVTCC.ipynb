```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.classify import NaiveBayesClassifier

# Downloading resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('names')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# --- 1. Part of Speech (POS) Tagging ---

text = "Python is an amazing programming language for AI."
words = word_tokenize(text)
pos_tags = nltk.pos_tag(words)
print("POS Tags:")
print(pos_tags)
```

```python
# --- 2. Gender Prediction (Simple Classifier) ---

def gender_features(word):
    return {'last_letter': word[-1]}

from nltk.corpus import names
labeled_names = ([(name, 'male') for name in names.words('male.txt')] + 
                 [(name, 'female') for name in names.words('female.txt')])

# Feature Extraction

featuresets = [(gender_features(n), g) for (n, g) in labeled_names]
classifier = NaiveBayesClassifier.train(featuresets)

# Test Prediction

name_to_test = "Frank"
print(f"Prediction for '{name_to_test}': {classifier.classify(gender_features(name_to_test))}")
```

```python
# --- 3. Simple Sentiment Analysis ---

def word_feats(words):
    return dict([(word, True) for word in words])

positive_vocab = ['awesome', 'outstanding', 'fantastic', 'great', 'nice']
negative_vocab = ['bad', 'terrible', 'useless', 'hate']

pos_features = [(word_feats([word]), 'pos') for word in positive_vocab]
neg_features = [(word_feats([word]), 'neg') for word in negative_vocab]

train_set = pos_features + neg_features
sentiment_classifier = NaiveBayesClassifier.train(train_set)

# Test Sentiment

sentence = "The movie was fantastic"
sentence_words = word_tokenize(sentence.lower())
pos_count = 0
neg_count = 0

for w in sentence_words:
    res = sentiment_classifier.classify(word_feats([w]))
    if res == 'pos': pos_count += 1
    if res == 'neg': neg_count += 1

print(f"\nSentence: '{sentence}'")
print(f"Positive Score: {pos_count/len(sentence_words):.2f}")
print(f"Negative Score: {neg_count/len(sentence_words):.2f}")
```
