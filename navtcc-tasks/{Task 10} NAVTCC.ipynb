## 1. Objective

To create a Python script that converts distance from kilometers to miles using a predefined conversion factor and user-provided data.

---

## 2. Learning Outcomes

- **Input and Casting:** Taking user input and immediately wrapping it in `float()` for numerical processing.  
- **Constant Usage:** Defining a fixed conversion factor (`1 km ≈ 0.621371 miles`).  
- **Dynamic Calculation:** Multiplying a variable by a constant to produce a real-time result.  
- **String Formatting:** Using `%0.2f` to ensure the distance is displayed with exactly two decimal places.  

---

## 3. Conclusion

The task confirms that Python can handle basic unit conversion logic, which is a critical skill for normalizing datasets and feature engineering in Machine Learning.

```python
kilometers = float(input("Enter value in kilometers: "))

conv_fac = 0.621371

miles = kilometers * conv_fac
print('%0.2f kilometers is equal to %0.2f miles' % (kilometers, miles))
```
