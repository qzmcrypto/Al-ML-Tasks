## **1. Objective**

The objective of this task is to understand the concepts of **Standard Deviation** and **Variance**. These metrics are crucial in **Machine Learning** for understanding the **spread of data** and measuring how much values **deviate from the mean**.

---

## **2. Learning Outcomes**

* **Standard Deviation (σ):** Understanding that this number describes the **spread of values**. A low value means data is close to the mean, while a high value indicates it is more spread out.

* **Variance (σ²):** Learning that variance is the **average of the squared differences** from the mean.

* **Mathematical Relationship:** Recognizing that **Standard Deviation** is the square root of **Variance** (`σ = √Variance`).

* **NumPy Integration:** Using `numpy.std()` for standard deviation and `numpy.var()` for variance to perform quick calculations on large datasets.

* **Data Interpretation:** Learning to **visualize data consistency**; for example, consistent car speeds have a low standard deviation, while inconsistent speeds have a high one.

---

## **3. Conclusion**

**Standard Deviation** and **Variance** are essential for **data preprocessing** and **model evaluation**. In AI, these values help identify **outliers** and understand the **stability of features**. Datasets with very high variance can make it harder for models to find patterns. Mastering these concepts allows us to **scale and normalize data** effectively before training **Neural Networks**.

```python
import numpy

# Case 1: Consistent speeds (Low spread)
speed1 = [86, 87, 88, 86, 87, 85, 86]

# Case 2: Inconsistent speeds (High spread)
speed2 = [32, 111, 138, 28, 59, 77, 97]
```

```python
# --- 1. Standard Deviation (std) ---
std_low = numpy.std(speed1)
std_high = numpy.std(speed2)

print(f"Low Standard Deviation: {std_low:.2f}")
print(f"High Standard Deviation: {std_high:.2f}")
```

```python
# --- 2. Variance (var) ---

var_high = numpy.var(speed2)
print(f"\nVariance of Case 2: {var_high:.2f}")
```

```python
# --- 3. Verification ---
# Square root of variance should equal standard deviation

print(f"Square root of Variance: {numpy.sqrt(var_high):.2f}")
```
