## **2. Learning Outcomes**

* **Mean (Average):** Learning to calculate the average value by summing all observations and dividing by the total count using `numpy.mean()`.

* **Median (Middle):** Understanding how to find the midpoint of a sorted dataset using `numpy.median()`, including how to handle both odd and even numbers of data points.

* **Mode (Common):** Using `stats.mode()` from the **SciPy** library to identify the most frequently occurring value in a dataset.

* **Statistical Application:** Recognizing that these three values provide a quick **mathematical snapshot** of a group of numbers, which is essential for **data preprocessing in AI**.

---

## **3. Conclusion**

**Mean, Median, and Mode** are the building blocks of data analysis. In Machine Learning, we use these techniques to **understand data distribution**, **fill in missing values (imputation)**, and **detect outliers**. Understanding these concepts ensures that we can **describe our data accurately** before feeding it into complex algorithms.

```python
import numpy
from scipy import stats

# Dataset: speed of 13 cars
speed = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]
```

```python
# --- 1. Calculate Mean ---
# The average value

mean_val = numpy.mean(speed)
print(f"Mean Speed: {mean_val}")
```

```python
# --- 2. Calculate Median ---
# The middle value after sorting

median_val = numpy.median(speed)
print(f"Median Speed: {median_val}")
```

```python
# --- 3. Calculate Mode ---
# The most common value

mode_val = stats.mode(speed)
print(f"Mode Speed: {mode_val.mode}")
print(f"Occurrence Count: {mode_val.count}")
```
