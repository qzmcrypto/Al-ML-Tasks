## 3. Conclusion

The task confirms that since the condition `20 < 15` is False, the program skips the if block and executes the else block instead, ensuring that one of the two paths is always followed before continuing with the rest of the script.

```python
i = 20
if (i < 15):
    print("i is smaller than 15")
    print("i'm in if Block")
else:
    print("i is greater than 15")
    print("i'm in else Block")

print("i'm not in if and not in else Block")
```
