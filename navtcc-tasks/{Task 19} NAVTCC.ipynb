```python
# --- 1. Using enumerate() for Index and Value ---

print("Using enumerate():")
words = ['The', 'Big', 'Bang', 'Theory']
for index, value in enumerate(words):
    print(index, value)
```

```python
# --- 2. Using zip() for Parallel Iteration ---

print("Using zip() for names and ages:")
names = ['Leo', 'Kendall', 'Harry']
ages = (24, 27, 25)
for name, age in zip(names, ages):
    print(f"Name: {name}, Age: {age}")
```

```python
# --- 3. Using items() for Dictionaries ---

print("Using items() for Dictionary:")
p1 = {'Washington': 'First', 'Lincoln': 'Emancipator', 'Roosevelt': 'Trust-Buster'}
for name, title in p1.items():
    print(f"{name}: {title}")
```

```python
# --- 4. Using sorted() and reversed() ---

num = [1, 3, 5, 6, 2, 1, 3]
print("Sorted list:")
for n in sorted(num):
    print(n, end=" ")

print("\nReversed list:")
for n in reversed(num):
    print(n, end=" ")
```

```python
# --- 5. While Loop with If Condition ---

print("While Loop with Condition:")
count = 0
while count < 5:
    if count == 3:
        print("Special case: Count is 3")
    count += 1
```
