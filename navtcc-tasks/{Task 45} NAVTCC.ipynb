## **2. Learning Outcomes**

* **Loading Data:** Using `pd.read_csv()` to import data into the Python environment.

* **Full Data Display:** Using the `to_string()` method to print the entire DataFrame without truncation.

* **System Options:** Checking the system's limit for displaying rows using `pd.options.display.max_rows`.

* **Modifying Settings:** Learning how to increase the display limit to view thousands of rows directly in the console.

## **3. Conclusion**

Mastering **CSV reading** is a core skill in **Data Science** because most datasets (from Kaggle, government databases, etc.) are provided in this format. Knowing how to **manipulate display options** allows an AI engineer to quickly **inspect large datasets** for patterns or errors before feeding the data into a **training pipeline**.

```python
import pandas as pd

# --- 1. Loading a CSV into a DataFrame ---
# (Note: Requires 'data.csv' in the same directory)
# For this task, we assume the file is available

df = pd.read_csv('data.csv')
```

```python
# --- 2. Printing the Entire DataFrame ---
# to_string() ensures nothing is hidden

print("--- Full Dataset (to_string) ---")
print(df.to_string()) 
```

```python
# --- 3. Default Print ---
# If to_string() isn't used, Pandas shows only first 5 and last 5 rows

print("--- Default Print (Truncated) ---")
print(df)
```

```python
# --- 4. Checking and Changing Max Rows ---
# Check the current system limit

print("Max Rows System Limit:", pd.options.display.max_rows) 

# Increase the limit to 9999

pd.options.display.max_rows = 9999

print("\n--- Print after changing max_rows ---")
print(df)
```
