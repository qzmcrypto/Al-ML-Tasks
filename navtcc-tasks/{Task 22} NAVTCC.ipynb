## **1. Objective**

The objective of this task is to understand **Python Tuples**. While similar to lists, tuples are distinct because they are **immutable** (unchangeable). This task covers how to create tuples, access elements via indexing, verify their immutability, and perform basic operations like finding length and iteration.

---

## **2. Learning Outcomes**

- **Tuple Characteristics:** - **Ordered:** They maintain the sequence of elements.  
  - **Immutable:** Once created, elements cannot be added, removed, or modified.  
  - **Duplicates:** Like lists, they allow multiple occurrences of the same value.

- **Indexing and Accessing:** - Using positive indices (starting from `0`) to retrieve specific data points.  
  - Understanding the internal structure where each item is mapped to a numerical index.

- **Immutability Constraints:** - Recognizing that assignment operations (e.g., `tuple[0] = value`) result in errors.  
  - Learning the "workaround" for modification: converting to a list and then back to a tuple.

- **Utility Functions:** - Using `len()` to count elements.  
  - Using `for` loops to process each item within the tuple.

---

## **3. Conclusion**

Tuples are essential when we want to ensure that our data remains constant throughout the program's execution, providing a layer of security and integrity. In AI and Machine Learning, tuples are frequently used to return multiple values from a function or to store fixed configurations.

```python
# --- 1. Creating a Python Tuple ---

# A tuple of numbers

numbers = (1, 2, -5)
print("Numbers Tuple:", numbers)

# Tuples can store different data types

mixed_tuple = ("Python", 101, True)
print("\nMixed Tuple:", mixed_tuple)
```

```python
# --- 2. Accessing Tuple Elements ---

languages = ('Python', 'Swift', 'C++')

# Accessing elements using index

print("First element (index 0):", languages[0])
print("\nThird element (index 2):", languages[2])
```

```python
# --- 3. Proving Immutability (The Error Case) ---

cars = ('BMW', 'Tesla', 'Ford', 'Toyota')

print("Trying to modify index 0...")
try:
    cars[0] = 'Nissan'
except TypeError as e:
    print(f"Error caught: {e} (Tuples cannot be changed!)")
```

```python
# --- 4. Tuple Length and Iteration ---

fruits = ('apple', 'banana', 'orange')

# Finding length

print("Total Items in fruits:", len(fruits))

# Iterating through the tuple

print("\nIterating through the tuple:")
for fruit in fruits:
    print(f"- {fruit}")
```

```python
# --- 5. Checking if an Item Exists ---

colors = ('Red', 'Green', 'Blue')
print("Is 'Red' in colors?", 'Red' in colors)
print("\nIs 'Yellow' in colors?", 'Yellow' in colors)
```
