## **1. Objective**

The objective of this task is to learn how to **Iterate** through NumPy arrays. Iterating means going through elements one by one. We will explore basic `for` loops for multi-dimensional arrays, the advanced `nditer()` function for efficient scalar iteration, and `ndenumerate()` for tracking indices during iteration.

---

## **2. Learning Outcomes**

* **Basic Iteration:** Understanding that iterating over a multi-dimensional array with a basic `for` loop moves through the first dimension (e.g., rows in a 2-D array, matrices in a 3-D array).

* **Nested Loops:** Learning that to reach individual (scalar) elements, multiple nested loops are required for high-dimensional arrays, which can become complex.

* **nditer() Helper:** Mastering the `nditer()` function, which allows us to visit every scalar element in an array without writing multiple nested loops, regardless of the array's dimensions.

* **Advanced nditer Flags:**
    * **Buffered:** Using `flags=['buffered']` and `op_dtypes` to change data types on the fly during iteration.
    * **ndenumerate():** Using this method to retrieve both the **index** and the **value** of elements simultaneously.

---

## **3. Conclusion**

While NumPy's strength lies in **vectorized operations** (which avoid manual loops for speed), iteration is sometimes necessary for specific custom logic or debugging. Using `nditer()` and `ndenumerate()` makes your code more readable and efficient by handling the complexity of multi-dimensional shapes automatically, which is crucial for coordinate-based data processing in AI.

```python
import numpy as np

# --- 1. Basic Iteration (1-D, 2-D, 3-D) ---
arr_1d = np.array([1, 2, 3])
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])

print("Iterating 1-D array:")
for x in arr_1d:
    print(x)

print("\nIterating 2-D (returns rows):")
for row in arr_2d:
    print(row)

# To get scalars from 2-D, we need nested loops
print("\nIterating 2-D (scalars):")
for row in arr_2d:
    for val in row:
        print(val)


# --- 2. Using nditer() for Efficient Iteration ---
# No matter how many dimensions, nditer visits every scalar
arr_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

print("\nUsing nditer() on 3-D array:")
for x in np.nditer(arr_3d):
    print(x)


# --- 3. nditer with Different Data Types ---
# Using buffered flag to iterate integers as strings
print("\nIterating as String type (S):")
for x in np.nditer(arr_1d, flags=['buffered'], op_dtypes=['S']):
    print(x)


# --- 4. Iterating with Different Step Size ---
print("\nIterating 2-D with step [:, ::2]:")
arr_step = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
for x in np.nditer(arr_step[:, ::2]):
    print(x)


# --- 5. Enumerated Iteration (ndenumerate) ---
print("\nEnumerated 2-D Iteration (Index, Value):")
for idx, x in np.ndenumerate(arr_step):
    print(f"Index: {idx}, Value: {x}")


# --- 6. Exercise Solution ---
# Given: arr = np.array(['a', 'b', 'c'])
# for x in arr: print('Hello ' + x)
ex_arr = np.array(['a', 'b', 'c'])
print("\nExercise Result:")
for x in ex_arr:
    print('Hello ' + x)
# Result: 
# Hello a
# Hello b
# Hello c
```
