## **1. Objective**

The objective of this task is to understand the functionality of the `break` statement in Python. We will demonstrate how it is used to exit a loop (both `for` and `while`) prematurely when a specific condition is met, shifting control to the next line of code outside the loop.

---

## **2. Learning Outcomes**

- **Premature Loop Exit:**  
  Learning how to stop a loop before it completes all iterations or before the condition naturally becomes `False`.  

- **Conditional Search:**  
  Implementing `break` to stop searching through a sequence once the target value is found.  

- **Scope in Nested Loops:**  
  Understanding that a `break` statement only exits the innermost loop it is placed in.  

- **Control Flow Management:**  
  Managing multiple exits in nested loops using flags (like boolean variables) alongside the `break` statement.  

---

## **3. Conclusion**

This task confirms that the `break` statement is a powerful jump statement that enhances the efficiency of our code. By exiting loops early, we save computational resources, which is crucial when processing large datasets in Machine Learning models.

```python
# --- Example 1: Break in a For Loop (Searching) ---

print("Searching for value 7 in a list:")
a = [1, 3, 5, 7, 9, 11]
val = 7

for i in a:
    if i == val:
        print(f"Found {i}! Exiting loop.")
        break
else:
    print("Value not found.")
```

```python
# --- Example 2: Break in a While Loop (Countdown) ---

print("Countdown using While True and Break:")
c = 5
while True:
    print(c)
    c -= 1
    if c == 0:
        print("Countdown finished!")
        break 
```

```python
# --- Example 3: Break in Nested Loops (Matrix Search) ---

print("Searching in a 2D Matrix (List of Lists):")
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
target = 5
found = False

for row in matrix:
    for num in row:
        if num == target:
            print(f"{target} found in matrix!")
            found = True
            break
    if found:
        break 
```
