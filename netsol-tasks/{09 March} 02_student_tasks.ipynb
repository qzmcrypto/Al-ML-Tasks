## ⭐ Task 1 — Calculate Basic Statistics

**Dataset:** Daily temperatures (°C) recorded for 10 days:
```python
temps = [22, 25, 19, 30, 27, 21, 28, 24, 26, 23]
```

**Your job:**
1. Calculate the **mean** temperature
2. Calculate the **median** temperature
3. Calculate the **mode** (most frequent temperature)
4. Print all three with clear labels

**Expected output:**
```
Mean   temperature: 24.50 °C
Median temperature: 24.50 °C
Mode   temperature: (the most repeated value)
```

> 💡 **Hints:**
> - `np.mean(data)` for mean
> - `np.median(data)` for median
> - `stats.mode(data).mode` for mode
> - Use `f'value: {x:.2f}'` for 2 decimal places

````python
# Task 1 — Your code here
temps = [22, 25, 19, 30, 27, 21, 28, 24, 26, 23]

# 1. Mean

mean_temp = np.mean(temps)

# 2. Median

median_temp = np.median(temps)

# 3. Mode

count = Counter(temps)
max_count = max(count.values())

if max_count == 1:
    mode_temp = "No mode (all values unique)"
else:
    mode_values = [k for k, v in count.items() if v == max_count]
    mode_temp = mode_values[0]

# 4. Print results

print(f"Mean   temperature: {mean_temp:.2f} °C")
print(f"Median temperature: {median_temp:.2f} °C")
print(f"Mode   temperature: {mode_temp}")
````

## ⭐ Task 2 — Measure the Spread

**Same dataset:**
```python
temps = [22, 25, 19, 30, 27, 21, 28, 24, 26, 23]
```

**Your job:**
1. Calculate the **range** (max − min)
2. Calculate the **variance**
3. Calculate the **standard deviation**
4. Print a sentence like: `'Most days are within ±X.XX°C of the average'`

**Expected output:**
```
Range   : 11
Variance: 10.25
Std Dev : 3.20
Most days are within ±3.20°C of the average (24.50°C)
```

> 💡 **Hints:**
> - Range = `max(data) - min(data)`
> - `np.var(data)` for variance
> - `np.std(data)` for standard deviation
> - Std dev is just `√variance` — try: `np.sqrt(np.var(data))`

````python
# Task 2 — Your code here
temps = [22, 25, 19, 30, 27, 21, 28, 24, 26, 23]

# 1. Range

data_range = max(temps) - min(temps)
print(f'Min:   {min(temps)}')
print(f'Max:   {max(temps)}')
print(f'Range: {data_range}')
print()

# 2. Variance

mean_temp = np.mean(temps)
variance_temp = np.var(temps, ddof=0)
print(f'Mean:     {mean_temp:.2f}°C')
print(f'Variance: {variance_temp:.2f}')
print()

# 3. Standard Deviation

std_dev_temp = np.sqrt(variance_temp)
print(f"Standard Deviation: {std_dev_temp:.2f}")

# 4. Print summary sentence

print(f"Most days are within ±{std_dev_temp:.2f}°C of the average ({mean_temp:.2f}°C)")
````

## ⭐ Task 3 — Quartiles & Five-Number Summary

**Dataset:** Marks of 12 students:
```python
marks = [45, 52, 58, 61, 67, 72, 76, 80, 85, 91, 95, 99]
```

**Your job:**
1. Find **Q1**, **Q2 (median)**, **Q3**
2. Calculate the **IQR** (Q3 − Q1)
3. Print the full **five-number summary** (Min, Q1, Q2, Q3, Max)

**Expected output:**
```
=== Five-Number Summary ===
Min    : 45
Q1     : 60.25
Median : 74.0
Q3     : 85.75
Max    : 99
IQR    : 25.5
```

> 💡 **Hints:**
> - `np.percentile(data, 25)` → Q1
> - `np.percentile(data, 50)` → Q2
> - `np.percentile(data, 75)` → Q3
> - IQR = Q3 − Q1

````python
# Task 3 — Your code here
marks = [45, 52, 58, 61, 67, 72, 76, 80, 85, 91, 95, 99]

# Q1, Q2, Q3

Q1 = np.percentile(marks, 25)
Q2 = np.percentile(marks, 50)
Q3 = np.percentile(marks, 75)

# IQR

IQR = Q3 - Q1

# Five-number summary

Min = min(marks)
Max = max(marks)

# Print results

print("=== Five-Number Summary ===")
print(f"Min    : {Min}")
print(f"Q1     : {Q1}")
print(f"Median : {Q2}")
print(f"Q3     : {Q3}")
print(f"Max    : {Max}")
print(f"IQR    : {IQR}")
````

## ⭐ Task 4 — Detect Outliers Using IQR

**Dataset:** Annual salaries at a small company (one person is the CEO!):
```python
salaries = [30000, 32000, 35000, 33000, 200000, 31000, 34000, 36000, 29000, 32500]
```

**Your job:**
1. Compute Q1, Q3, and IQR
2. Compute the **outlier fences**:
   - Lower fence = Q1 − 1.5 × IQR
   - Upper fence = Q3 + 1.5 × IQR
3. Loop through salaries and **print any outliers**

**Expected output:**
```
Lower fence : 22,875.00
Upper fence : 43,625.00
🚨 Outlier detected: 200,000
```

> 💡 **Hints:**
> - Use `for s in salaries:` loop
> - Check: `if s < lower or s > upper:`
> - 200,000 (the CEO salary) should be the outlier
> - Use `f'{s:,}'` to format numbers with commas

````python
# Task 4 — Your code here
salaries = [30000, 32000, 35000, 33000, 200000,
            31000, 34000, 36000, 29000, 32500]

# Step 1: Compute Q1, Q3, and IQR

Q1 = np.percentile(salaries, 25)
Q3 = np.percentile(salaries, 75)
IQR = Q3 - Q1

# Step 2: Compute outlier fences

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Step 3: Print the fences

print("=== Outlier Fences ===")
print(f"Lower fence : {lower_fence:,.2f}")
print(f"Upper fence : {upper_fence:,.2f}")

# Step 4: Detect and print outliers

print("\n=== Outliers Detected ===")
for s in salaries:
    if s < lower_fence or s > upper_fence:
        print(f"🚨 Outlier detected: {s:,}")
````

## ⭐⭐ Task 5 — Z-Score Normalisation

**Dataset:** Heights (cm) of 10 people:
```python
heights = [150, 162, 170, 155, 180, 175, 168, 158, 172, 165]
```

**Your job:**
1. Compute the **mean** and **std** of heights
2. Calculate the **z-score** for each height
3. Print each height with its z-score
4. Flag anyone with **|z| > 1.5** as `'⚠️ Unusual'`

**Expected output:**
```
Height  Z-Score  Flag
 150     -1.84   ⚠️ Unusual
 162     -0.68
 170     +0.09
 ...
 180     +1.08
```

> 💡 **Hints:**
> - `z = (x - mean) / std`
> - Loop: `for h in heights:`
> - `flag = '⚠️ Unusual' if abs(z) > 1.5 else ''`
> - This is exactly what `StandardScaler` does in scikit-learn!

````python
# Task 5 — Your code here
heights = [150, 162, 170, 155, 180, 175, 168, 158, 172, 165]

# Compute mean and std

mean_height = np.mean(heights)
std_height = np.std(heights)

# Print table header

print(f'  {"Height":>6}  {"Z-Score":>8}  Flag')
print('-' * 30)

# Calculate and print z-scores
# for h in heights:
#     z = ...
#     flag = ...
#     print(...)

for h in heights:
    z = (h - mean_height) / std_height
    flag = '⚠️ Unusual' if abs(z) > 1.5 else ''
    print(f'{h:6}  {z:8.2f}  {flag}')
````

## ⭐⭐ Task 6 — Explore Skewness

**Create these two datasets:**
```python
right_skewed = [2, 3, 3, 4, 4, 4, 5, 5, 10, 15, 20]
left_skewed  = [1, 5, 10, 15, 16, 16, 17, 17, 18, 18, 19]
```

**Your job:**
1. Compute **skewness** for both datasets using `stats.skew()`
2. Compute **mean** and **median** for both
3. Print results and verify the rule:
   - Right skew → mean **>** median
   - Left skew  → mean **<** median

**Expected output:**
```
--- Right Skewed ---
Skewness: +1.56   direction: positive →
Mean: 7.73   Median: 4.0
Mean > Median ✓  (right skew rule holds!)

--- Left Skewed ---
Skewness: -1.38   direction: negative ←
Mean: 13.0   Median: 16.0
Mean < Median ✓  (left skew rule holds!)
```

> 💡 **Hints:**
> - `stats.skew(data)` returns a number
> - Positive = right skew, Negative = left skew
> - Check the mean vs median relationship each time

````python
# Task 6 — Your code here
right_skewed = [2, 3, 3, 4, 4, 4, 5, 5, 10, 15, 20]
left_skewed  = [1, 5, 10, 15, 16, 16, 17, 17, 18, 18, 19]

# For right_skewed:

mean_right = np.mean(right_skewed)
median_right = np.median(right_skewed)
skew_right = stats.skew(right_skewed)
direction_right = "positive →" if skew_right > 0 else "negative ←"
rule_right = "✓" if mean_right > median_right else "✗"

# For left_skewed:

mean_left = np.mean(left_skewed)
median_left = np.median(left_skewed)
skew_left = stats.skew(left_skewed)
direction_left = "positive →" if skew_left > 0 else "negative ←"
rule_left = "✓" if mean_left < median_left else "✗"

# Print and verify the skew rule

print(f"""--- Right Skewed ---
Skewness: {skew_right:+.2f}   direction: {direction_right}
Mean: {mean_right:.2f}   Median: {median_right}
Mean > Median {rule_right}  (right skew rule holds!)

--- Left Skewed ---
Skewness: {skew_left:+.2f}   direction: {direction_left}
Mean: {mean_left:.2f}   Median: {median_left}
Mean < Median {rule_left}  (left skew rule holds!)
""")
````

## ⭐⭐ Task 7 — Correlation Analysis

**Dataset:** Hours of sleep and test scores for 10 students:
```python
hours_sleep = [5, 6, 6, 7, 7, 8, 8, 9, 9, 10]
test_score  = [55, 60, 62, 68, 71, 75, 78, 82, 85, 88]
```

**Your job:**
1. Calculate the **Pearson correlation** and **p-value**
2. Determine strength: Strong (>0.8) / Moderate (>0.5) / Weak
3. State if the result is statistically significant (p < 0.05)
4. **Bonus:** Plot a scatter plot with a trend line

**Expected output:**
```
Pearson r = 0.997
p-value   = 0.000001
Strength : STRONG positive correlation
Significant: YES ✅
```

> 💡 **Hints:**
> - `corr, pval = stats.pearsonr(x, y)`
> - Use `if/elif/else` for strength labels
> - Bonus: `plt.scatter(x, y)` then `np.polyfit(x, y, 1)` for trend line

````python
# Task 7 — Your code here
hours_sleep = [5, 6, 6, 7, 7, 8, 8, 9, 9, 10]
test_score  = [55, 60, 62, 68, 71, 75, 78, 82, 85, 88]

# 1. Pearson correlation

corr, pval = stats.pearsonr(hours_sleep, test_score)

# 2. Strength classification

if abs(corr) > 0.8:
    strength = "STRONG"
elif abs(corr) > 0.5:
    strength = "MODERATE"
else:
    strength = "WEAK"

# 3. Significance check

significant = "YES ✅" if pval < 0.05 else "NO ❌"

# 4. Print

print(f"Pearson r = {corr:.3f}")
print(f"p-value   = {pval:.6f}")
print(f"Strength : {strength} positive correlation" if corr > 0 else f"Strength : {strength} negative correlation")
print(f"Significant: {significant}")

# 5. (Bonus) Scatter plot

plt.scatter(hours_sleep, test_score)
m, b = np.polyfit(hours_sleep, test_score, 1)
plt.plot(hours_sleep, [m*x + b for x in hours_sleep], color='red')
plt.xlabel("Hours of Sleep")
plt.ylabel("Test Score")
plt.title("Sleep vs Test Score")
plt.show()
````

## ⭐⭐⭐ Task 9 — Hypothesis Testing

**Scenario:** A school tested a new teaching method.

```python
old_method = [65, 70, 68, 72, 74, 69, 71, 73, 66, 75]
new_method = [78, 82, 80, 85, 79, 83, 81, 84, 77, 86]
```

**Your job:**
1. State **H₀** and **H₁** in plain English (as print statements)
2. Run an **independent samples t-test**
3. Print t-statistic and p-value
4. Print the means of both groups
5. Write a clear **conclusion** based on the p-value

**Expected output:**
```
H₀: Both methods produce the same results
H₁: The new method is significantly different

Old method mean: 70.3
New method mean: 81.5
t-statistic    : -10.470
p-value        : 0.000000

✅ Reject H₀ — new method is significantly better!
```

> 💡 **Hints:**
> - `stats.ttest_ind(group1, group2)`
> - If p < 0.05 → reject H₀
> - Compare the means to see which group performed better
> - The t-stat being negative just means group1 < group2

````python
# Task 9 — Your code here
old_method = [65, 70, 68, 72, 74, 69, 71, 73, 66, 75]
new_method = [78, 82, 80, 85, 79, 83, 81, 84, 77, 86]

# 1. State hypotheses

print("H₀: Both methods produce the same results")
print("H₁: The new method is significantly different\n")

# 2. Run t-test

t_stat, p_val = stats.ttest_ind(old_method, new_method)

# 3. Print t-stat and p-value

print(f"t-statistic    : {t_stat:.3f}")
print(f"p-value        : {p_val:.6f}\n")

# 4. Print group means

mean_old = sum(old_method)/len(old_method)
mean_new = sum(new_method)/len(new_method)
print(f"Old method mean: {mean_old:.1f}")
print(f"New method mean: {mean_new:.1f}\n")

# 5. Conclusion

if p_val < 0.05:
    if mean_new > mean_old:
        print("✅ Reject H₀ — new method is significantly better!")
    else:
        print("✅ Reject H₀ — old method is significantly better!")
else:
    print("❌ Fail to reject H₀ — no significant difference")
````

## ⭐⭐⭐ Task 10 — MINI PROJECT: Build a Stats Report Function 🏆

**Write a function `stats_report(data, name)` that prints a complete statistical summary.**

It should output:
1. Dataset name and values
2. Mean, Median, Mode
3. Std Dev and Variance
4. Q1, Q2, Q3 and IQR
5. Skewness with direction label
6. Any outliers (using IQR method)
7. Z-scores — flag any value where |z| > 2

**Test it with:**
```python

```

**Expected output:**
```
========================================
STATS REPORT: Sample Data

--- Central Tendency ---
```

> 💡 **Hints:**
> - Use `def stats_report(data, name):`
> - Call all the numpy/scipy functions inside the function
> - Use `stats.mode(data).mode` and `.count` for mode
> - Build it step by step — get one section working, then add the next

> 🚀 **Bonus Challenges:**
> - Make the function **return a dictionary** instead of just printing
> - Add a **histogram** chart of the data
> - Accept a **pandas Series** as input
> - Test it on the `scores` and `salaries` datasets from earlier tasks

> 🎯 If you can build this function from scratch, you truly understand statistics for ML!

````python
# Task 10 — Your code here
# Build your stats_report function below

def stats_report(data, name):
    print('=' * 40)
    print(f'STATS REPORT: {name}')
    print('=' * 40)
    
    # 1. Values

    print(f"Values    : {data}\n")
    
    # 2. Central Tendency (mean, median, mode)

    mean = np.mean(data)
    median = np.median(data)
    mode_result = stats.mode(data, keepdims=True)
    mode_val = mode_result.mode[0]
    mode_count = mode_result.count[0]
    
    print("--- Central Tendency ---")
    print(f"Mean      : {mean:.2f}")
    print(f"Median    : {median:.2f}")
    print(f"Mode      : {mode_val} (appeared {mode_count}x)\n")

    # 3. Spread (std, variance)

    std_dev = np.std(data, ddof=0)
    variance = np.var(data, ddof=0)
    
    print("--- Spread ---")
    print(f"Std Dev   : {std_dev:.2f}")
    print(f"Variance  : {variance:.2f}\n")
    
    # 4. Quartiles (Q1, Q2, Q3, IQR)

    Q1 = np.percentile(data, 25)
    Q2 = np.percentile(data, 50)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    
    print("--- Quartiles ---")
    print(f"Q1={Q1:.2f}  Q2={Q2:.2f}  Q3={Q3:.2f}  IQR={IQR:.2f}\n")

    # 5. Skewness

    skewness = stats.skew(data)
    if skewness > 0:
        direction = "right skewed"
    elif skewness < 0:
        direction = "left skewed"
    else:
        direction = "symmetric"   
        
    print("--- Shape ---")
    print(f"Skewness  : {skewness:+.2f}  ({direction})\n")
    
    # 6. Outliers

    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    outliers = [x for x in data if x < lower or x > upper]
    
    print("--- Outliers (IQR method) ---")
    if outliers:
        for val in outliers:
            print(f"🚨 Outlier: {val}")
    else:
        print("No outliers detected")
    print()
    
    # 7. Z-scores

    print("--- Z-Scores ---")
    z_scores = (np.array(data) - mean)/std_dev
    for val, z in zip(data, z_scores):
        flag = "⚠️ EXTREME" if abs(z) > 2 else ""
        print(f" {val:3}   z={z:+.2f} {flag}")
    
    print('=' * 40)

# Test it!
stats_report([12, 15, 14, 10, 100, 13, 14, 15, 16, 11], 'Sample Data')
````
